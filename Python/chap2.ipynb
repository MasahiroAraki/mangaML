## ロジスティック識別

breast_cancer データに対してロジスティック識別の学習を行い、識別関数の重みを表示します。LogisticRegression の繰り返し回数がデフォルトでは 100回と少なかったので、10000回に変更しています。

```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression

breast_cancer = load_breast_cancer()
X = breast_cancer.data
y = breast_cancer.target

clf1 = LogisticRegression(max_iter=10000)
clf1.fit(X,y)
for f, w in zip(breast_cancer.feature_names, clf1.coef_[0]) :
    print("{0:<23}: {1:6.2f}". format(f, w))
```

# 決定木学習

scikit-learn ver 0.21以降に導入された [plot_tree](https://scikit-learn.org/stable/modules/generated/sklearn.tree.plot_tree.html)関数によって、学習結果の決定木の表示が出版社の公式ページで配布されているコードよりも簡単になりました。

```python
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

breast_cancer = load_breast_cancer()
X = breast_cancer.data
y = breast_cancer.target

clf2 = DecisionTreeClassifier(max_depth=2)
clf2.fit(X, y)
tree.plot_tree(clf2,
               feature_names=breast_cancer.feature_names,
               class_names=breast_cancer.target_names,
               filled=True,
               fontsize=9)
plt.show()
```
